```python
import json

def count_records_in_jsonl(file_path):
    count = 0

    with open(file_path, 'r') as file:
        for line in file:
            count += 1

    return count

# Provide the path to your JSONL file
jsonl_file_path = 'dataset/test.jsonl'

record_count = count_records_in_jsonl(jsonl_file_path)
print("Total number of records:", record_count)
```
